# **> Encoding** -- **Get_Dummies**:

```python
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True)
df1
```

**> Concatenate**

```python
data=pd.concat([df,df1],axis=1)
data
```

**> Removing unwanted columns**

```python
data=data.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],axis=1)
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

**> Removing string values from columns (Mileage,Engine,Power)**

```python
#Mileage

data['Mileage']=data['Mileage'].str.replace('km/kg','')
data
```

```python
data['Mileage']=data['Mileage'].str.replace('kmpl','')
data
```

```python
#Engine

data['Engine']=data['Engine'].str.replace('CC','')
data
```

```python
#Power

data['Power']=data['Power'].str.replace('bhp','')
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

**>  Replacing 'null' values by '0**

```python
data['Mileage']=data['Mileage'].str.replace('null','0')
data['Power']=data['Power'].str.replace('null','0')
data['Engine']=data['Engine'].str.replace('null','0')
```

```python
data.isna().sum()
```

**> Converting objects into numerical type (float/int)**

```python
#Mileage,Engine,Power

data['Mileage']=data['Mileage'].astype(float)
data['Engine']=data['Engine'].astype(float)
data['Power']=data['Power'].astype(float)
```

```python
data.dtypes
```

**> Converting given '0' value for 'null' to Missing value**

```python
data.loc[data.Mileage==0,'Mileage']=np.NaN
data.loc[data.Engine==0,'Engine']=np.NaN
data.loc[data.Power==0,'Power']=np.NaN
```

```python
data.isna().sum()
```

```python
#null value converted to missing value..So missing value increased
```

**> Filling Missing value**

```python
lst=['Mileage','Power','Engine']
for i in lst:
   data[i]=data[i].fillna(data[i].mean())
data['Seats']=data['Seats'].fillna(data['Seats'].mode()[0])
```

```python
data.isna().sum()
```

```python
sns.heatmap(data.corr())
```

**> Independent variable**

```python
x=data.drop(['Price'],axis=1)
x
```

**> Dependent variable**

```python
y=data['Price']
y
```
